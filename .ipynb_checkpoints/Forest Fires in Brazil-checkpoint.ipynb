```python
df['month_eng'] = df['month'].map({'Janeiro': 'Jan',
                                   'Fevereiro': 'Feb',
                                   'Mar√ßo': 'Mar', 
                                   'Abril': 'April', 
                                   'Maio': 'May',
                                   'Junho': 'Jun',
                                   'Julho': 'July',
                                   'Agosto': 'Aug',
                                   'Setembro':'Sep',
                                   'Outubro': 'Oct',
                                   'Novembro':'Nov',
                                   'Dezembro': 'Dec'})
df.head()
```

```python
df.nunique()
```

```python
monthlyFire = df.groupby('month_eng', as_index= False)['number'].sum()
monthlyFire
```

```python
plt.figure(figsize= (15,5))
sns.barplot(data= monthlyFire, x= 'month_eng', y= 'number')
```
